```python
def turnover_situation(df, station, time, time_window, rolling_stock, num_carriages, driver_change):
    
    # Convert 'time' string to a time object
    target_time = pd.to_datetime(time, format='%H:%M:%S').time()
    
    # Extract only the time part from PLAN_DATETIME
    df['time_only'] = df['PLAN_DATETIME'].dt.time
    
    # Create a function to check if the time is within the time window
    def within_time_window(row_time):
        # Convert times to timedelta since midnight for comparison
        row_time_td = pd.Timedelta(hours=row_time.hour, minutes=row_time.minute, seconds=row_time.second)
        target_time_td = pd.Timedelta(hours=target_time.hour, minutes=target_time.minute, seconds=target_time.second)
        # Calculate the difference in minutes
        time_diff = abs((row_time_td - target_time_td).total_seconds() / 60)
        return time_diff <= time_window
    
    # Apply the time window filter
    df = df[df['time_only'].apply(within_time_window)]
    
    # Apply other filters
    filtered_df = df[
        (df['STATION'] == station) &
        (df['ROLLINGSTOCK_TYPE'] == rolling_stock) &
        (df['NUMBER_CARRIAGES'] == num_carriages) &
        (df['DRIVER_CHANGE'] == driver_change)
    ]
    
    # Remove the auxiliary column before returning
    filtered_df.drop(columns=['time_only'], inplace=True)
    
    return filtered_df
```

```python
result_df = turnover_situation(turnover_df, 'Ut', '08:00:00', 60, 'ICM', 6, '1.0')
```
