```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
df_large = pd.read_csv("large_preprocessed_week_24-05-2024.csv", low_memory=False)
df_large = df_large.drop(df_large.columns[0], axis=1)
```

```python
df = df_large
df = df[df["REALIZED_TURNOVER_TIME"] < 6000]
```

```python
temp_df = df.copy()
temp_df = temp_df[temp_df["REALIZED_TURNOVER_TIME"] < 6000]

# Create the distribution plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=temp_df, x='REALIZED_TURNOVER_TIME', color='black')

# Add a vertical dashed line at 2500
plt.axvline(x=120, color='red', linestyle='--')
plt.axvline(x=1800, color='red', linestyle='--')

# Add text around the line, slightly to the left
x_text = 2400  
y_text = plt.gca().get_ylim()[1] * 0.9  

#plt.text(x_text, y_text, 'Filtering Point (2500)', color='red', ha='right')
# Set plot title and labels
plt.xlabel('Realized Turnover Time (seconds)')
plt.ylabel('Density')


# Show the plot
plt.show()
```

```python
# Copy the dataframe
temp_df = df.copy()
temp_df = temp_df[temp_df["REALIZED_TURNOVER_TIME"] < 1800]
temp_df = temp_df[temp_df["REALIZED_TURNOVER_TIME"] > 120]

temp_df['HOUR'] = temp_df['HOUR'].astype(str)
# Ensure 'Hour' column is ordered chronologically
temp_df['HOUR'] = pd.Categorical(temp_df['HOUR'], categories=[str(i) for i in range(24)], ordered=True)

# Plotting
plt.figure(figsize=(14, 8))
ax = sns.boxplot(x='HOUR', y='REALIZED_TURNOVER_TIME', data=temp_df, showfliers=False, 
                 boxprops=dict(facecolor='white', edgecolor='black'))

# Add titles and labels
plt.title('Realized Turnover Time For Each Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Realized Turnover Time (seconds)')
plt.tight_layout()
plt.show()
```

```python
temp_df = df.copy()
temp_df = temp_df[temp_df["REALIZED_TURNOVER_TIME"] < 1800]
temp_df = temp_df[temp_df["REALIZED_TURNOVER_TIME"] > 120]

# Convert 'DAY_OF_WEEK' to string
temp_df['DAY_OF_WEEK'] = temp_df['DAY_OF_WEEK'].astype(str)

# Map numeric values to days of the week
day_mapping = {str(i): day for i, day in enumerate(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])}
temp_df['DAY_OF_WEEK'] = temp_df['DAY_OF_WEEK'].map(day_mapping)

# Ensure 'DAY_OF_WEEK' column is ordered chronologically
temp_df['DAY_OF_WEEK'] = pd.Categorical(temp_df['DAY_OF_WEEK'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)

# Plotting
plt.figure(figsize=(14, 8))
ax = sns.boxplot(x='DAY_OF_WEEK', y='REALIZED_TURNOVER_TIME', data=temp_df, showfliers=False, 
                 boxprops=dict(facecolor='white', edgecolor='black'))

# Add titles and labels
plt.title('Realized Turnover Time For Each Day')
plt.xlabel('Day of the Week')
plt.ylabel('Realized Turnover Time (seconds)')
plt.tight_layout()
plt.show()
```
